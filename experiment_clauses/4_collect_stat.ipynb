```python
subset = False

if subset:
    CLAUSES_DB_FILE = "./clauses_trees_experiment_subset_of_5000_20221113.db"
else:
    CLAUSES_DB_FILE = "./clauses_trees_experiment_20221113.db"
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(CLAUSES_DB_FILE) 
conn.row_factory = sqlite3.Row  
cursor = conn.cursor()
```

```python
cursor.execute('SELECT count(*) as totalrows FROM clauses ')
totalrows =  cursor.fetchone()
totalrows['totalrows']
```

```python
cursor.execute('SELECT DISTINCT texttype FROM clauses ')
subcoropora = [r['texttype'] for r  in cursor.fetchall()]
subcoropora
```

**PÃ¤rime korpuste kaupa, muidu on liiga aeglane**

```python
if subset:
    filename_template = '5000_stat_subcorpora_occurrences_by_freq_%s.csv'
else:
    filename_template = 'stat_subcorpora_occurrences_by_freq_%s.csv'
    
for corpus in subcoropora:
    
    Query = f"""
    SELECT  total AS freq, count(*) AS  occurrences FROM clauses 
    WHERE texttype='{corpus}'
    group by   total
    """

    dfTotal = pd.read_sql_query(Query, conn)
    sum = dfTotal['occurrences'].sum()
    dfTotal['percent_of_all_occurrences'] = round(dfTotal['occurrences'] / sum * 100,3)
    if subset:
        dfTotal.to_csv(filename_template%corpus, index=None)
   
    print(f'--------------- {corpus} top 50  ----------')
    print(dfTotal.head(50))
    print('')
```
