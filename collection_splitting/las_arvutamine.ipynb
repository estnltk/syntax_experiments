```python
from estnltk.taggers.standard.syntax.syntax_error_count_tagger import by_sentences
```

```python
def get_las_score(layer_orig, layer_short):
    lases = []
    for spans, deprels, LASes, UASes, LAs in by_sentences(layer_short, layer_orig):
        if spans is not None:
            # Labelled Attachment Score (LAS)
            # Unlabelled Attachment Score (UAS)  round(UASes.count(0)*100/len(UASes),1)
            # Label Accuracy (LA)  round(LAs.count(0)*100/len(LAs),1)
           
            print(LASes)
            print("samad servad %", round(LASes.count(0)*100/len(LASes),1))
            
            #return  round(LASes.count(0)*100/len(LASes),1),  round(UASes.count(0)*100/len(UASes),1),  round(LAs.count(0)*100/len(LAs),1)  
        #return None, None, None 
```

```python
# lühike ja pikk

get_las_score(long.stanza_syntax, long.syntax_without_entity_advmod )
```

### pikalt läbi 

```python
def las2(layer_orig, layer_short):
    
    short_edges = []
    for span in layer_short:
        if span.children != ():
            for ch in span.children:
                short_edges.append((span.text, ch.text, ch.deprel))
        if span.deprel == "root":
            short_edges.append((None, span.text, span.deprel))
    print("lühikese lause servad:\n",short_edges)
    
    
    long_edges = []
    for span in layer_orig:
        if span.children != ():
            for ch in span.children:
                long_edges.append((span.text, ch.text, ch.deprel))            
        if span.deprel == "root":
            long_edges.append((None, span.text, span.deprel))
    print("pika lause servad:\n", long_edges)
    
    same = []
    for edge in short_edges:
        print(edge, "on pikas:", edge in long_edges)
        same.append(edge in long_edges)
    print(same, "\n", "samad servad %", round(same.count(True)*100/len(same), 1))
```

```python
las2(long.stanza_syntax, long.syntax_without_entity_advmod )
```
