### Fraasidele vastavate Text-objektide väljavõtmine

```python
text_obl_phrases = []
text_nsubj_phrases = []
text_obj_phrases = []
```

```python
test_text.obl_phrases[0].text
```

```python
for i in range(len(test_text.obl_phrases)):
    # create string from word list
    text_obj = " ".join(test_text.obl_phrases[i].text)
    # create Text-object and tag necessary layers
    text_obj = Text(text_obj).tag_layer(['morph_analysis', 'timexes'])
    # add metadata
    text_obj.meta['phrase_nr'] = i
    text_obj.meta['phrase_type'] = 'obl_phrase'
    # tag syntax analysis layer
    stanza_tagger.tag(text_obj)
    # tag ner layer
    ner_tagger.tag(text_obj)
    # append to list
    text_obl_phrases.append(text_obj)
```

```python
for i in range(len(test_text.nsubj_phrases)):
    # create string from list
    text_obj = " ".join(test_text.nsubj_phrases[i].text)
    # create Text-object and tag necessary layers
    text_obj = Text(text_obj).tag_layer(['morph_analysis', 'timexes'])
    # add metadata
    text_obj.meta['phrase_nr'] = i
    text_obj.meta['phrase_type'] = 'nsubj_phrase'
    # tag syntax analysis layer
    stanza_tagger.tag(text_obj)
    # tag ner layer
    ner_tagger.tag(text_obj)
    # append to list
    text_nsubj_phrases.append(text_obj)
```

```python
for i in range(len(test_text.obj_phrases)):
    # create string from list
    text_obj = " ".join(test_text.obj_phrases[i].text)
    # create Text-object and tag necessary layers
    text_obj = Text(text_obj).tag_layer(['morph_analysis', 'timexes'])
    # add metadata
    text_obj.meta['phrase_nr'] = i
    text_obj.meta['phrase_type'] = 'obj_phrase'
    # tag syntax analysis layer
    stanza_tagger.tag(text_obj)
    # tag ner layer
    ner_tagger.tag(text_obj)
    # append to list
    text_obj_phrases.append(text_obj)
```

### Süntaksipuu graafi esitus

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
test_phrase.stanza_syntax
```

Graafi koostamisel on võetud eeskuju siit: https://github.com/estnltk/syntax_experiments/blob/subcat/obl_data_experiments/data_helpers/syntax_graph.py

```python
def create_graph(stanza_syntax_layer):
    graph = nx.DiGraph()
    
    for word in stanza_syntax_layer:
        if isinstance(word['id'], int):
            # graaf pannakse kokku
            graph.add_node(
                word['id'],
                id=word['id'],
                lemma=word['lemma'],
                pos=word['upostag'],
                deprel=word['deprel'],
                form=word.text,
                feats=word['feats'],
                start=word.start,
                end=word.end)
            graph.add_edge(word['id'] - word['id'] + word['head'], word['id'], deprel=word['deprel'])
            
    return graph
```

```python
def draw_graph(graph):
    plt.rcParams["figure.figsize"] = (18.5, 10.5)
    pos = graphviz_layout(graph, prog='dot')
    labels = nx.get_node_attributes(graph, 'lemma')
    nx.draw(graph, pos, cmap=plt.get_cmap('jet'), labels=labels, with_labels=True, node_color='red')
    edge_labels = nx.get_edge_attributes(graph, 'deprel')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels)
    plt.show()
    plt.clf()
```

```python
test_graph = create_graph(test_phrase.stanza_syntax)
```

```python
test_graph.nodes(data=True)
```

```python
list(test_graph.nodes(data=True))
```

```python
# Visualiseerimine
draw_graph(test_graph)
```

```python
all_graphs = []
for text in df['text']:
    graph = create_graph(text.stanza_syntax)
    all_graphs.append(graph)
df.insert(len(df.columns), 'graph', all_graphs)
```
