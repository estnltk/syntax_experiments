```python
from estnltk.corpus_processing.parse_koondkorpus import get_div_target
from estnltk.corpus_processing.parse_koondkorpus import parse_tei_corpus

from estnltk import Text

from estnltk_neural.taggers import StanzaSyntaxTagger
from estnltk.taggers.standard.syntax.phrase_extraction.phrase_extractor import PhraseExtractor

#from estnltk_core.layer_operations import split_by
#from estnltk_core.layer_operations import split_by_sentences

from estnltk.taggers import NerTagger
```

```python
# sisendfail
input_xml_file = "C:\\Users\\liivas\\Documents\\Töö\\Postimees\\Postimees\\PM\\postimees_1995\\aja_pm_1995_12_07.xml"
```

```python
target = get_div_target( input_xml_file )
```

```python
texts = []
```

```python
# import documents as Text objects   
for text_obj in parse_tei_corpus( input_xml_file, target=[target] ):
    texts.append(text_obj)
```

```python
len(texts)
```

```python
# Alustuseks võetakse failist pikim tekst
test_text = texts[0]

for i in range(1, len(texts)):
    if len(texts[i].text) > len(test_text.text):
        test_text = texts[i]
```

### Stanza süntaksianalüüsikihi märgendamine

```python
# Tagger
stanza_tagger = StanzaSyntaxTagger(input_type='morph_analysis', input_morph_layer='morph_analysis',
                                   add_parent_and_children=True)
```

```python
stanza_tagger.tag(test_text)
```

### PhraseExtractor nimisõnafraaside leidmiseks

```python
# fraasid, mille peasõnaks on obliikva
obl_phrase_extractor = PhraseExtractor(deprel="obl", syntax_layer="stanza_syntax",
                                   output_layer="obl_phrases")

# fraasid, mille peasõnaks on nimisõnast subjekt
nsubj_phrase_extractor = PhraseExtractor(deprel="nsubj", syntax_layer="stanza_syntax",
                                   output_layer="nsubj_phrases")

# fraasid, mille peasõnaks on objekt
obj_phrase_extractor = PhraseExtractor(deprel="obj", syntax_layer="stanza_syntax",
                                   output_layer="obj_phrases")
```

```python
obl_phrase_extractor.tag( test_text )
nsubj_phrase_extractor.tag( test_text )
obj_phrase_extractor.tag( test_text )
```

### Fraasidele vastavate Text-objektide väljavõtmine

```python
text_obl_phrases = []
text_nsubj_phrases = []
text_obj_phrases = []
```

```python
test_text.obl_phrases[0].text
```

```python
for i in range(len(test_text.obl_phrases)):
    # create string from word list
    text_obj = " ".join(test_text.obl_phrases[i].text)
    # create Text-object and tag necessary layers
    text_obj = Text(text_obj).tag_layer(['morph_analysis', 'timexes'])
    # add metadata
    text_obj.meta['phrase_type'] = 'obl_phrase'
    # tag syntax analysis layer
    stanza_tagger.tag(text_obj)
    # tag ner layer
    ner_tagger.tag(text_obj)
    # append to list
    text_obl_phrases.append(text_obj)
```

```python
for i in range(len(test_text.nsubj_phrases)):
    # create string from list
    text_obj = " ".join(test_text.nsubj_phrases[i].text)
    # create Text-object and tag necessary layers
    text_obj = Text(text_obj).tag_layer(['morph_analysis', 'timexes'])
    # add metadata
    text_obj.meta['phrase_type'] = 'nsubj_phrase'
    # tag syntax analysis layer
    stanza_tagger.tag(text_obj)
    # tag ner layer
    ner_tagger.tag(text_obj)
    # append to list
    text_nsubj_phrases.append(text_obj)
```

```python
for i in range(len(test_text.obj_phrases)):
    # create string from list
    text_obj = " ".join(test_text.obj_phrases[i].text)
    # create Text-object and tag necessary layers
    text_obj = Text(text_obj).tag_layer(['morph_analysis', 'timexes'])
    # add metadata
    text_obj.meta['phrase_type'] = 'obj_phrase'
    # tag syntax analysis layer
    stanza_tagger.tag(text_obj)
    # tag ner layer
    ner_tagger.tag(text_obj)
    # append to list
    text_obj_phrases.append(text_obj)
```
