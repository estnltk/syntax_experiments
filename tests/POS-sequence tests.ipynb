```python
from estnltk import Text
from POS_sequence import get_POS_sequence
```

```python
# tekst üheste sõnaliigimärgenditega
text1 = Text("Ilus suur kollane kass").tag_layer('morph_analysis')
text1.morph_analysis
```

```python
# tekst on tühi sõne
text2 = Text("").tag_layer('morph_analysis')
text2.morph_analysis
```

```python
# tekst koosneb tühikutest
text3 = Text("  ").tag_layer('morph_analysis')
text3.morph_analysis
```

```python
# tekstis mitmene sõnaliigimärgend
text4 = Text("Hoolega tehtud töö").tag_layer('morph_analysis')
text4.morph_analysis
```

```python
# teksti lõpus tühik
text5 = Text("Hoolega tehtud töö ").tag_layer('morph_analysis')
text5.morph_analysis
```

```python
# tekst koosneb ühest sõnast
text6 = Text("Kass").tag_layer('morph_analysis')
text6.morph_analysis
```

```python
# tekst koosneb ühest sõnast, millel on kaks analüüsi identsete sõnaliigimärgenditega
text7 = Text("Varahoidlasse").tag_layer('morph_analysis')
text7.morph_analysis
```

```python
assert get_POS_sequence(text1) == 'A-A-A-S', get_POS_sequence(text1)
```

```python
assert get_POS_sequence(text2) == '', get_POS_sequence(text2)
```

```python
assert get_POS_sequence(text3) == '', get_POS_sequence(text3)
```

```python
assert get_POS_sequence(text4) == 'S-(A|V)-S', get_POS_sequence(text4)
```

```python
assert get_POS_sequence(text5) == 'S-(A|V)-S', get_POS_sequence(text5)
```

```python
assert get_POS_sequence(text6) == 'S', get_POS_sequence(text6)
```

```python
assert get_POS_sequence(text7) == 'S', get_POS_sequence(text7)
```
