```python
def tree_to_markdown(text_obj):
    markdown = ''
    for i in range(len(text_obj.stanza_syntax)):
        stanza_word = text_obj.stanza_syntax[i]
        ner = None
        nertag = None
        
        if len(text_obj.ner) > 0:
            word = text_obj.words.get(stanza_word)
            for n in text_obj.ner:
                for part in n:
                    if part==word:
                        ner=word
                        nertag=n.nertag
                        
        timex = text_obj.timexes.get(stanza_word)
        if ner:
            markdown += f"{stanza_word.id}\t{nertag}\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n"
        elif timex:
            markdown += f"{stanza_word.id}\tTIMEX\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n"
        else:
            markdown += f"{stanza_word.id}\tOTHER\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n"
    return markdown
```

```python
def df_to_markdown(df):
    markdown = ''
    for _, row in df.iterrows():
        markdown += f"## {row['pos_sequence']}\n\nTree:\n\n{row['tree']}\nhas_ner: {row['has_ner']}\n\nhas_timex: {row['has_timex']}\n\ncount_of_tree: {row['count_of_tree']}\n\n%_trees_same_POS: {row['%_trees_same_pos']}\n\n%_all_trees: {row['%_all_trees']}\n\n"
    return markdown
```

```python
# creating data
data = {'pos_sequence': [], 'tree': [], 'has_ner': [], 'has_timex': [], 'count_of_tree': [], 
        '%_trees_same_len': [], '%_all_trees': []}

data['pos_sequence'].append(test_phrase['pos_sequence_verb_info'])
data['tree'].append(tree_to_markdown(test_phrase['phrase']))
data['has_ner'].append(test_phrase['has_ner_netity'])
data['has_timex'].append(test_phrase['has_timex_entity'])
data['count_of_tree'].append(1)
data['%_trees_same_len'].append(1/1)
data['%_all_trees'].append(round((1/100)*100, 2))
        
data_df = pd.DataFrame(data)
md_content = df_to_markdown(data_df)  
print(md_content)
```

Ainult fraasidel, milles esineb ner/timex:

```python
n_phrases = len(df)

# creating data
data = {'pos_sequence': [], 'tree': [], 'has_ner': [], 'has_timex': [], 'count_of_tree': [], 
        '%_trees_same_pos': [], '%_all_trees': []}

for name, group in grouped_df:
    #group = sklearn.utils.shuffle(group).reset_index(drop=True)
    grouped_2 = group.groupby('graph_code_pos_ner_timex_verb_info')
    for name2, group2 in grouped_2:
        group2 = sklearn.utils.shuffle(group2).reset_index(drop=True)
        
        data['pos_sequence'].append(name)
        data['tree'].append(tree_to_markdown(group2['phrase'][0]))
        data['has_ner'].append('ner_' in name2)
        data['has_timex'].append('timex' in name2)
        data['count_of_tree'].append(len(group2))
        data['%_trees_same_pos'].append(round((len(group2)/len(group))*100, 2))
        data['%_all_trees'].append(round((len(group2)/n_phrases)*100, 2))
        
data_df = pd.DataFrame(data)
md_content = df_to_markdown(data_df)  
print(md_content)
with open('phrase_examples_filtered/complete/ner_timex_trees.md', 'w') as f:
    f.write(md_content)
```

```python
grouped_df_top50 = df_top50.groupby('pos_sequence_verb_info')

n_phrases = len(df_top50)

# creating data
data = {'pos_sequence': [], 'tree': [], 'has_ner': [], 'has_timex': [], 'count_of_tree': [], 
        '%_trees_same_pos': [], '%_all_trees': []}

for name, group in grouped_df_top50:
    #group = sklearn.utils.shuffle(group).reset_index(drop=True)
    if name in pos_sequences:
        grouped_2 = group.groupby('graph_code_pos_ner_timex_verb_info')
        for name2, group2 in grouped_2:
            if 'ner_' in name2 or 'timex' in group2:
                group2 = sklearn.utils.shuffle(group2).reset_index(drop=True)
        
                data['pos_sequence'].append(name)
                data['tree'].append(tree_to_markdown(group2['phrase'][0]))
                data['has_ner'].append('ner_' in name2)
                data['has_timex'].append('timex' in name2)
                data['count_of_tree'].append(len(group2))
                data['%_trees_same_pos'].append(round((len(group2)/len(group))*100, 2))
                data['%_all_trees'].append(round((len(group2)/n_phrases)*100, 2))
        
data_df = pd.DataFrame(data)
md_content = df_to_markdown(data_df)  
print(md_content)
with open('phrase_examples_filtered/complete/ner_timex_trees_top50.md', 'w') as f:
    f.write(md_content)
```
