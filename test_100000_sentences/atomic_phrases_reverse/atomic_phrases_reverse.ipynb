```python
# function for deciding, whether given phrase is atomic or not
def is_atomic(phrase):
    root = None
    for i in range(len(phrase.stanza_syntax)):
        if phrase.stanza_syntax[i]['deprel'] == 'root':
            root = phrase.stanza_syntax[i]

    if root.children:
        for child in root.children:
            # if root's direct child has children and direct child is noun modifier, current phrase is not atomic
            if child.children and child.deprel == 'nmod':
                return False
    return True
```

```python
def df_to_markdown1(df):
    markdown = ''
    for _, row in df.iterrows():
        markdown += f"## Tree: {row['tree']}\n\nPOS-sequence: {row['pos_sequence']}\n\nExample phrase: *{row['example_phrase']}*\n\nFrequency: {row['count']}\n\n% of all filtered phrases: {row['%_all_atomic_phrases']}\n\nNumber of unique syntax trees (POS-tags): {row['n_unique_trees_pos']}\n\nNumber of unique syntax trees (POS-tags and NER/TIMEX): {row['n_unique_trees_pos_ner_timex']}\n\n% of trees occurring once (POS-tags): {row['%_trees_pos_occurring_once']}\n\n% of trees occurring once (POS-tags and NER/TIMEX): {row['%_trees_pos_ner_timex_occurring_once']}\n\n"
    return markdown
```

```python
n_more_frequent = 0       
            
df = pd.read_pickle('phrase_examples_filtered_sub10000/atomic_phrases.pkl')

to_be_kept = [True]*len(df)
for idx, row in df.iterrows():
    if row['phrase_length'] < 2:
        to_be_kept[idx] = False

filtered_df = df[to_be_kept]
            
grouped_df = filtered_df.groupby('graph_code_verb_info')
# finding 50 most common POS-sequences
sorted_groups = grouped_df.size().reset_index(name='count').sort_values(['count'], ascending=False)

top50_trees = dict()
idx = 0
for i, row in sorted_groups.iterrows():
    top50_trees[row['graph_code_verb_info']] = row['count']
    idx+=1
    if idx == 50:
        break
        
# total number of phrases in DataFrame
n_phrases = len(df)
        
# creating data
data = {'tree': [], 'pos_sequence': [], 'example_phrase': [], 'count': [], '%_all_atomic_phrases': [], 
        'n_unique_trees_pos': [], 'n_unique_trees_pos_ner_timex': [], 
        '%_trees_pos_occurring_once': [], '%_trees_pos_ner_timex_occurring_once': []}

# new DataFrame for top50 trees
new_df = pd.DataFrame()

for name, group in grouped_df:
    if name in top50_trees.keys():
        if top50_trees[name] > 1:
            n_more_frequent+=1
            
            new_df = pd.concat([new_df, group], ignore_index=True)
            
            grouped_pos = group.groupby('pos_sequence_verb_info')   
            
            for name2, group2 in grouped_pos:
                shuffled_group = sklearn.utils.shuffle(group2).reset_index(drop=True)
                data['tree'].append(name)
                data['pos_sequence'].append(name2)
                data['example_phrase'].append(shuffled_group['phrase'][0].text)
                data['count'].append(top50_trees[name])
                data['%_all_atomic_phrases'].append(round((top50_trees[name]/n_phrases)*100, 2))
        
                data['n_unique_trees_pos'].append(len(group2['graph_code_pos_verb_info'].unique()))
                data['n_unique_trees_pos_ner_timex'].append(len(group2['graph_code_pos_ner_timex_verb_info'].unique()))
        
        
                n_all_trees_pos = len(group2['graph_code_pos_verb_info'])
                counts_pos = group2['graph_code_pos_verb_info'].value_counts().tolist()
                n_single_occurrences_pos = 0
                for count in counts_pos:
                    if count == 1:
                        n_single_occurrences_pos+=1
                
                n_all_trees_pos_ner_timex = len(group2['graph_code_pos_ner_timex_verb_info'])
                n_single_occurrences_pos_ner_timex = 0
                counts_pos_ner_timex = group2['graph_code_pos_ner_timex_verb_info'].value_counts().tolist()
                for count in counts_pos_ner_timex:
                    if count == 1:
                        n_single_occurrences_pos_ner_timex+=1
        
                data['%_trees_pos_occurring_once'].append(round((n_single_occurrences_pos/n_all_trees_pos)*100, 2))
                data['%_trees_pos_ner_timex_occurring_once'].append(round((n_single_occurrences_pos_ner_timex/n_all_trees_pos_ner_timex)*100, 2))
        
new_df.to_pickle('phrase_examples_filtered_sub10000/atomic_phrases_reverse_top50.pkl')

data_df = pd.DataFrame(data)
data_df = data_df.sort_values('count', ascending=False)
md_content = df_to_markdown1(data_df)
with open('phrase_examples_filtered_sub10000/atomic_phrases_reverse_top50.md', 'w', encoding='utf-8') as f:
    f.write(md_content)
```

```python
def tree_to_markdown(text_obj):
    markdown = ''
    for i in range(len(text_obj.stanza_syntax)):
        stanza_word = text_obj.stanza_syntax[i]
        ner = None
        nertag = None
        
        if len(text_obj.ner) > 0:
            word = text_obj.words.get(stanza_word)
            for n in text_obj.ner:
                for part in n:
                    if part==word:
                        ner=word
                        nertag=n.nertag
                        
        timex = text_obj.timexes.get(stanza_word)
        if ner:
            markdown += f"{stanza_word.id}\t{nertag}\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n\n"
        elif timex:
            markdown += f"{stanza_word.id}\tTIMEX\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n\n"
        else:
            markdown += f"{stanza_word.id}\tOTHER\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n\n"
    return markdown
```

```python
def get_subphrases(phrase, root, sub_phrases=[]):
    POS_tags = ['S', 'Y', 'P', 'H']
    
    if root.upostag not in POS_tags:
        if len(root.children)>0:
            for child in root.children:
                return get_subphrases(phrase, child, sub_phrases)
        else:
            return sub_phrases
        
    elif root.upostag in POS_tags:
        if len(root.children)>0:
            subphrase = []
            for i in range(len(phrase.stanza_syntax)):
                if phrase.stanza_syntax[i] == root or phrase.stanza_syntax[i] in root.children and phrase.stanza_syntax[i].upostag != 'H':
                    subphrase.append(phrase.stanza_syntax[i])
                elif phrase.stanza_syntax[i] == root or phrase.stanza_syntax[i] in root.children and phrase.stanza_syntax[i].upostag == 'H':
                    subphrase.append(phrase.stanza_syntax[i])
                    if len(phrase.stanza_syntax[i].children)>0:
                        for subchild in phrase.stanza_syntax[i].children:
                            if subchild.upostag == 'H' and subchild.deprel == 'flat':
                                subphrase.append(subchild)
                                
            for child in root.children:
                return get_subphrases(phrase, child, sub_phrases + [subphrase])
            #return get_subphrases(phrase, child, sub_phrases + [[root]])
        
        else:
            return sub_phrases + [[root]]
        
    #elif root.upostag == 'H':
    #    if len(root.children)>0:
    #        subphrase = []
    #        for i in range(len(phrase.stanza_syntax)):
    #            if phrase.stanza_syntax[i] == root or phrase.stanza_syntax[i] in root.children:
    #                subphrase.append(phrase.stanza_syntax[i])
    #        for child in root.children:
    #            if child.upostag == 'H' and child.deprel == 'flat' and len(child.children)>0:
    #                return get_subphrases(phrase, child, sub_phrases + [subphrase])
    #            elif child.upostag == 'H' and child.deprel == 'flat' and len(child.children)==0:
    #                return sub_phrases + [subphrase]
    #            else:
    #                return get_subphrases(phrase, child, sub_phrases + [[root]] + [subphrase])
    #    else:
    #        return sub_phrases + [[root]]
            
```
