```python
# function for deciding, whether given phrase is atomic or not
def is_atomic(phrase):
    root = None
    for i in range(len(phrase.stanza_syntax)):
        if phrase.stanza_syntax[i]['deprel'] == 'root':
            root = phrase.stanza_syntax[i]

    if root.children:
        for child in root.children:
            # if root's direct child has children and direct child is noun modifier, current phrase is not atomic
            if child.children and child.deprel == 'nmod':
                return False
    return True
```

```python
def tree_to_markdown(text_obj):
    markdown = ''
    for i in range(len(text_obj.stanza_syntax)):
        stanza_word = text_obj.stanza_syntax[i]
        ner = None
        nertag = None
        
        if len(text_obj.ner) > 0:
            word = text_obj.words.get(stanza_word)
            for n in text_obj.ner:
                for part in n:
                    if part==word:
                        ner=word
                        nertag=n.nertag
                        
        timex = text_obj.timexes.get(stanza_word)
        if ner:
            markdown += f"{stanza_word.id}\t{nertag}\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n\n"
        elif timex:
            markdown += f"{stanza_word.id}\tTIMEX\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n\n"
        else:
            markdown += f"{stanza_word.id}\tOTHER\t{stanza_word.text}\t{stanza_word.head}\t{stanza_word.deprel}\n\n"
    return markdown
```

```python
def df_to_markdown3(df):
    markdown = ''
    for _, row in df.iterrows():
        markdown += f"## {row['pos_sequence']}\n\nTree:\n\n{row['tree']}\nhas_ner: {row['has_ner']}\n\nhas_timex: {row['has_timex']}\n\ncount_of_tree: {row['count_of_tree']}\n\n%_trees_same_POS: {row['%_trees_same_pos']}\n\n%_all_trees: {row['%_all_trees']}\n\n"
    return markdown
```

### Subphrases of non-atomic phrases

Võtame DataFrame-i eelnevalt leitud mitte-atomaarsete fraasidega ning rakendame nende peal meetodit alam(nimisõna)fraaside eraldamiseks.

```python
def get_subphrases(phrase, root, sub_phrases=[]):
    POS_tags = ['S', 'Y', 'P', 'H']
    
    if root.upostag not in POS_tags:
        if len(root.children)>0:
            for child in root.children:
                return get_subphrases(phrase, child, sub_phrases)
        else:
            return sub_phrases
        
    elif root.upostag in POS_tags:
        if len(root.children)>0:
            subphrase = []
            for i in range(len(phrase.stanza_syntax)):
                if phrase.stanza_syntax[i] == root or phrase.stanza_syntax[i] in root.children and phrase.stanza_syntax[i].upostag != 'H':
                    subphrase.append(phrase.stanza_syntax[i])
                elif phrase.stanza_syntax[i] == root or phrase.stanza_syntax[i] in root.children and phrase.stanza_syntax[i].upostag == 'H':
                    subphrase.append(phrase.stanza_syntax[i])
                    if len(phrase.stanza_syntax[i].children)>0:
                        for subchild in phrase.stanza_syntax[i].children:
                            if subchild.upostag == 'H' and subchild.deprel == 'flat':
                                subphrase.append(subchild)
                                
            for child in root.children:
                return get_subphrases(phrase, child, sub_phrases + [subphrase])
            #return get_subphrases(phrase, child, sub_phrases + [[root]])
        
        else:
            return sub_phrases + [[root]]
        
    #elif root.upostag == 'H':
    #    if len(root.children)>0:
    #        subphrase = []
    #        for i in range(len(phrase.stanza_syntax)):
    #            if phrase.stanza_syntax[i] == root or phrase.stanza_syntax[i] in root.children:
    #                subphrase.append(phrase.stanza_syntax[i])
    #        for child in root.children:
    #            if child.upostag == 'H' and child.deprel == 'flat' and len(child.children)>0:
    #                return get_subphrases(phrase, child, sub_phrases + [subphrase])
    #            elif child.upostag == 'H' and child.deprel == 'flat' and len(child.children)==0:
    #                return sub_phrases + [subphrase]
    #            else:
    #                return get_subphrases(phrase, child, sub_phrases + [[root]] + [subphrase])
    #    else:
    #        return sub_phrases + [[root]]
            
```
